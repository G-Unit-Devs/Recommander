```python
import os
import json
import pandas as pd

# üìÇ D√©finir le chemin du repo
repo_path = os.getcwd()

# üìã Lister tous les fichiers JSON
json_files = []
for root, dirs, files in os.walk(repo_path):
    for file in files:
        if file.endswith(".json"):
            json_files.append(os.path.join(root, file))  # Ajouter le chemin complet

# üìå Charger et fusionner tous les fichiers JSON
all_data = []

for json_file in json_files:
    with open(json_file, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)  # Charger le fichier JSON
            if isinstance(data, list):  # Si c'est une liste de profils
                for entry in data:
                    entry["source"] = json_file  # Ajouter une colonne source
                    all_data.append(entry)
            elif isinstance(data, dict):  # Si c'est un seul objet JSON
                data["source"] = json_file
                all_data.append(data)
        except json.JSONDecodeError:
            print(f"‚ö†Ô∏è Erreur dans le fichier {json_file}")

# üìä Convertir en DataFrame
full_df = pd.DataFrame(all_data)


```

```python
import ast

# üìå V√©rifier et convertir "data" en dictionnaire si n√©cessaire
def parse_json_safe(value):
    if isinstance(value, str):  # V√©rifier si c'est une cha√Æne JSON mal interpr√©t√©e
        try:
            return ast.literal_eval(value)  # Convertir string en dict
        except (ValueError, SyntaxError):
            return {}  # Retourner un dictionnaire vide en cas d'erreur
    return value  # Si d√©j√† un dict, on le garde tel quel

# Appliquer la conversion sur la colonne "data"
full_df["data"] = full_df["data"].apply(parse_json_safe)

# üìå Fonction pour extraire une cl√© sp√©cifique de "data"
def extract_from_data(column, key):
    if isinstance(column, dict):  # V√©rifier si la valeur est bien un dictionnaire
        return column.get(key, "")  # R√©cup√©rer la valeur de la cl√© demand√©e
    return ""

# Appliquer l'extraction des informations utiles
full_df["expectations"] = full_df["data"].apply(lambda x: extract_from_data(x, "expectations"))
full_df["domain"] = full_df["data"].apply(lambda x: extract_from_data(x, "domain"))
full_df["experience"] = full_df["data"].apply(lambda x: extract_from_data(x, "experience"))
full_df["about"] = full_df["data"].apply(lambda x: extract_from_data(x, "about"))


print(full_df.head())  # Afficher les 5 premi√®res lignes

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# üìå Fonction pour convertir une valeur en string
def to_string(value):
    if isinstance(value, list):  # Si c'est une liste, convertir en texte
        return " ".join(map(str, value))
    elif pd.isna(value):  # Si c'est NaN, remplacer par une cha√Æne vide
        return ""
    else:  # Sinon, convertir normalement en string
        return str(value)

# üìå Extraire les champs "data" pour les chercheurs et pros
def extract_and_clean(df, col_name):
    df[col_name] = df["data"].apply(lambda x: to_string(x.get(col_name, "")) if isinstance(x, dict) else "")
    return df

# S√©parer chercheurs et pros
chercheurs_df = full_df[full_df["role"] == "chercheur"].copy()
pros_df = full_df[full_df["role"] == "pro"].copy()

# üìå Liste des crit√®res √† comparer
criteres = ["expectations", "domain", "experience", "about", "help", "passions", "histoires"]

# Appliquer l'extraction et nettoyage sur toutes les colonnes
for critere in criteres:
    chercheurs_df = extract_and_clean(chercheurs_df, critere)
    pros_df = extract_and_clean(pros_df, critere)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# üìå V√©rifier et nettoyer les colonnes de texte
def clean_text_column(df, col_name):
    if col_name in df.columns:
        df[col_name] = df[col_name].fillna("").astype(str)  # Remplace NaN par "" et force en str
        df[col_name] = df[col_name].apply(lambda x: x.strip())  # Supprime les espaces inutiles
    return df

# üìå V√©rifier et corriger les chercheurs et pros pour chaque crit√®re
criteres = ["expectations", "domain", "experience", "about", "help", "passions", "histoires"]

for critere in criteres:
    chercheurs_df = clean_text_column(chercheurs_df, critere)
    pros_df = clean_text_column(pros_df, critere)

# üìå Initialisation d'un dictionnaire pour stocker les r√©sultats
similarity_results = {}

# üìå Appliquer TF-IDF et calculer la similarit√© pour chaque crit√®re
for critere in criteres:
    print(f"üîç Traitement du crit√®re : {critere}")  # Ajout de logs pour le suivi

    vectorizer = TfidfVectorizer()
    
    # V√©rification si la colonne contient des valeurs valides
    if chercheurs_df[critere].str.strip().eq("").all() or pros_df[critere].str.strip().eq("").all():
        print(f"‚ö†Ô∏è Le crit√®re '{critere}' est vide pour tous les chercheurs ou pros. Il est ignor√©.")
        continue  # Passer au crit√®re suivant

    # Vectorisation des textes
    chercheurs_tfidf = vectorizer.fit_transform(chercheurs_df[critere])
    pros_tfidf = vectorizer.transform(pros_df[critere])

    # Calcul de la similarit√© cosinus
    similarity_matrix = cosine_similarity(chercheurs_tfidf, pros_tfidf)

    # Stocker la matrice dans le dictionnaire
    similarity_results[critere] = similarity_matrix

# üìå Transformer les r√©sultats en DataFrames pour affichage
similarity_dfs = {
    critere: pd.DataFrame(similarity_results[critere], index=chercheurs_df.index, columns=pros_df.index)
    for critere in similarity_results
}

# üéØ Afficher un aper√ßu de la similarit√© pour chaque crit√®re
import ace_tools as tools
for critere, df in similarity_dfs.items():
    tools.display_dataframe_to_user(name=f"Similarit√© pour {critere}", dataframe=df)
```

```python
# üìå Calculer l'importance de chaque crit√®re en faisant la moyenne des scores
importance_scores = {critere: similarity_results[critere].mean() for critere in criteres}

# üìä Transformer en DataFrame pour analyse
importance_df = pd.DataFrame.from_dict(importance_scores, orient='index', columns=["Poids Moyen"])
importance_df = importance_df.sort_values(by="Poids Moyen", ascending=False)

# üéØ Afficher les scores d'importance
tools.display_dataframe_to_user(name="Importance des Crit√®res", dataframe=importance_df)
```
